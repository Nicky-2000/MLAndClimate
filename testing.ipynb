```python
## Aggregating trips by vehicle ID
veh_id_results = {}
veh_id_weird_households = []

# Iterate through each household ID
for veh_id in vehicle_ids:
    
    vehicle_trips = trip2pub_filtered_used_car[trip2pub_filtered_used_car["VEHCASEID"] == veh_id] # Get all the trips for this car
    vehicle_trips = vehicle_trips.sort_values(by=["TDAYDATE", "STRTTIME"])  # Sort by date and time of the trips
    vehicle_trips["TRPHHVEH"].to_list()
    
    # Check that all trips are made in the household vehicle
    if not all(vehicle_trips["TRPHHVEH"] == 1): # SHOULD NEVER GET TRIGGERED since we filtered trips to only include the car trips
        print(f"Household ID {household_id} has trips not made in household vehicle")
        continue
    
    if vehicle_trips.empty:
        # This means the household made no trips (with their car at least) during this day
        travel_day = merged[merged["VEHCASEID"] == veh_id]["TRAVDAY_x"].values[0]
        results[veh_id] = {
            "trips": [], # No trips were made
            "household_id": household_id,
            "day": travel_day,
        }
        continue
    
    # Otherwise, we need to find all the times the vehicle left the house and returned

    current_departure = None
    current_day = None
    
    trips = []
    travel_day = merged[merged["VEHCASEID"] == veh_id]["TRAVDAY_x"].values[0]

    for idx, row in vehicle_trips.iterrows():
        whyfrom = row["WHYFROM"]
        whyto = row["WHYTO"]
        trip_start_time = row["STRTTIME"]
        trip_end_time = row["ENDTIME"]
        miles = row["VMT_MILE"]
        duration = row["TRVLCMIN"]
        day = row["TRAVDAY"]
        assert day == travel_day # should be the same day as the household travel day

        # Case 1: One-row round trip
        if whyfrom in [1, 2] and whyto in [1, 2]:
            trips.append((trip_start_time, trip_end_time, miles))
            current_departure = None  # Reset state in case something was set

        # Case 2: Leaving home (start of trip)
        elif whyfrom in [1, 2]:
            current_departure = trip_start_time

        # Case 3: Returning home (end of a trip)
        elif whyto in [1, 2] and current_departure is not None:
            trips.append((current_departure, trip_end_time, miles))
            current_departure = None

        # Case 4: Unexpected state (returning home without leaving)
        elif whyto in [1, 2] and current_departure is None:
            print(f"veh_id {veh_id} arrived home without prior departure â€” ignoring row")
    
    if current_departure is not None:
        # This means they left home but never returned
        print(f"veh_id {veh_id} left home but never returned")
        veh_id_weird_households.append(household_id)
        continue
    # Add the trips to the results
    veh_id_results[veh_id] = {
        "trips": trips, # No trips were made
        "day": travel_day,
        }

print(f"Total trips found: {len(veh_id_results)}")
print(f"Weird Households: {len(veh_id_weird_households)}")
```
