```python
trip2pub_filtered_used_car = trip2pub_filtered[trip2pub_filtered["TRPHHVEH"] == 1]
```

```python
vehicle_ids = merged["VEHCASEID"].to_list()
```

```python
## Aggregating trips by vehicle ID
veh_id_results = {}
veh_id_weird_households = []

# Iterate through each household ID
for veh_id in vehicle_ids:
    
    vehicle_trips = trip2pub_filtered_used_car[trip2pub_filtered_used_car["VEHCASEID"] == veh_id] # Get all the trips for this car
    vehicle_trips = vehicle_trips.sort_values(by=["TDAYDATE", "STRTTIME"])  # Sort by date and time of the trips
    vehicle_trips["TRPHHVEH"].to_list()
    
    # Check that all trips are made in the household vehicle
    if not all(vehicle_trips["TRPHHVEH"] == 1): # SHOULD NEVER GET TRIGGERED since we filtered trips to only include the car trips
        print(f"Household ID {household_id} has trips not made in household vehicle")
        continue
    
    if vehicle_trips.empty:
        # This means the household made no trips (with their car at least) during this day
        travel_day = merged[merged["VEHCASEID"] == veh_id]["TRAVDAY_x"].values[0]
        veh_id_results[veh_id] = {
            "trips": [], # No trips were made
            "day": travel_day,
        }
        continue
    
    # Otherwise, we need to find all the times the vehicle left the house and returned

    current_departure = None
    current_day = None
    
    trips = []
    travel_day = merged[merged["VEHCASEID"] == veh_id]["TRAVDAY_x"].values[0]
    total_trip_miles = 0
    for _, row in vehicle_trips.iterrows():
        whyfrom = row["WHYFROM"]
        whyto = row["WHYTO"]
        trip_start_time = row["STRTTIME"]
        trip_end_time = row["ENDTIME"]
        miles = row["VMT_MILE"]

        # Case 1: start and end at home in one trip
        if whyfrom in [1, 2] and whyto in [1, 2]:
            trips.append((trip_start_time, trip_end_time, miles))
            current_departure = None
            total_trip_miles = 0

        # Case 2: Leaving home...
        elif whyfrom in [1, 2]:
            current_departure = trip_start_time
            total_trip_miles = miles  # start accumulating miles

        # Case 3: Going from location1 -> location2 where neighter is home
        elif whyto not in [1, 2] and current_departure is not None:
            total_trip_miles += miles  # accumulate miles for this leg

        # Case 4: Returning home
        elif whyto in [1, 2] and current_departure is not None:
            total_trip_miles += miles  # include this last leg
            trips.append((current_departure, trip_end_time, total_trip_miles))
            current_departure = None
            total_trip_miles = 0

        # Case 5: Arrived home without a known departure
        elif whyto in [1, 2] and current_departure is None:
            print(f"veh_id {veh_id} arrived home without prior departure â€” ignoring row")
    
    if current_departure is not None:
        # This means they left home but never returned
        print(f"veh_id {veh_id} left home but never returned")
        veh_id_weird_households.append(household_id)
        continue
    # Add the trips to the results
    veh_id_results[veh_id] = {
        "trips": trips, # No trips were made
        "day": travel_day,
        }

print(f"Total trips found: {len(veh_id_results)}")
print(f"Weird Households: {len(veh_id_weird_households)}")
```

```python
household_ids = merged["HOUSEID"].to_list()
```

```python
## Aggregating trips by household
house_id_results = {}
house_id_weird_households = []

# Iterate through each household ID
for household_id in household_ids:
    
    household_trips = trip2pub_filtered_used_car[trip2pub_filtered_used_car["HOUSEID"] == household_id] # Get all the trips for this household (which should have 1 car)
    household_trips = household_trips.sort_values(by=["TDAYDATE", "STRTTIME"])  # Sort by date and time of the trips
    household_trips["TRPHHVEH"].to_list()
    
    # Check that all trips are made in the household vehicle
    if not all(household_trips["TRPHHVEH"] == 1): # SHOULD NEVER GET TRIGGERED since we filtered trips to only include the car trips
        print(f"Household ID {household_id} has trips not made in household vehicle")
        continue
    
    if household_trips.empty:
        # This means the household made no trips (with their car at least) during this day
        travel_day = merged[merged["HOUSEID"] == household_id]["TRAVDAY_x"].values[0]
        house_id_results[household_id] = {
            "trips": [], # No trips were made
            "household_id": household_id,
            "day": travel_day,
        }
        continue
    
    # Otherwise, we need to find all the times the vehicle left the house and returned

    current_departure = None
    current_day = None
    
    trips = []
    travel_day = merged[merged["HOUSEID"] == household_id]["TRAVDAY_x"].values[0]

    for idx, row in household_trips.iterrows():
        whyfrom = row["WHYFROM"]
        whyto = row["WHYTO"]
        trip_start_time = row["STRTTIME"]
        trip_end_time = row["ENDTIME"]
        miles = row["VMT_MILE"]
        duration = row["TRVLCMIN"]
        day = row["TRAVDAY"]
        assert day == travel_day # should be the same day as the household travel day

        # Case 1: One-row round trip
        if whyfrom in [1, 2] and whyto in [1, 2]:
            trips.append((trip_start_time, trip_end_time, miles))
            current_departure = None  # Reset state in case something was set

        # Case 2: Leaving home (start of trip)
        elif whyfrom in [1, 2]:
            current_departure = trip_start_time

        # Case 3: Returning home (end of a trip)
        elif whyto in [1, 2] and current_departure is not None:
            trips.append((current_departure, trip_end_time, miles))
            current_departure = None

        # Case 4: Unexpected state (returning home without leaving)
        elif whyto in [1, 2] and current_departure is None:
            print(f"Household {household_id} arrived home without prior departure â€” ignoring row")
    
    if current_departure is not None:
        # This means they left home but never returned
        print(f"Household ID {household_id} left home but never returned")
        house_id_weird_households.append(household_id)
        continue
    # Add the trips to the results
    house_id_results[household_id] = {
        "trips": trips, # No trips were made
        "day": travel_day,
        }

print(f"Total trips found: {len(house_id_results)}")
print(f"Weird Households: {len(house_id_weird_households)}")
```
